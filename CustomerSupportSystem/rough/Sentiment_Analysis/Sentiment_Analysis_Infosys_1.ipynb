```python
import os
import requests

# Get API key from environment variable
API_KEY = os.getenv('GROQ_API_KEY')
if not API_KEY:
    raise ValueError("API key not found. Set GROQ_API_KEY in your environment variables.")

# GROQ API endpoint
API_URL = " https://api.groq.com/openai/v1"

# Function to analyze sentiment
def analyze_sentiment(text):
    try:
        # Define headers and payload
        headers = {
            "Authorization": f"Bearer {API_KEY}",
            "Content-Type": "application/json"
        }
        payload = {
            "text": text  # Text input for sentiment analysis
        }

        # Make the API request
        response = requests.post(API_URL, json=payload, headers=headers)

        # Raise exception for HTTP errors
        response.raise_for_status()

        # Parse the JSON response
        data = response.json()

        # Extract sentiment from the response
        sentiment = data.get("sentiment", "Unknown")
        return sentiment
    except requests.exceptions.RequestException as e:
        return f"API Request Error: {e}"
    except Exception as e:
        return f"Error: {e}"

# Example usage
if __name__ == "__main__":
    text = "The product is amazing! I'm very happy with it."
    sentiment = analyze_sentiment(text)
    print(f"Sentiment: {sentiment}")
```

```python
import openai
from together import Together
import pandas as pd
import time
from sklearn.metrics import classification_report

# Initialize API clients
openai.api_key = open_ai_api  # Replace with your OpenAI API Key
together_client = Together(api_key=together_ai_api)

# Function for OpenAI sentiment analysis
def openai_sentiment(text):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",  # Adjust to the model available in your OpenAI account
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": f"Provide the sentiment (positive, negative, neutral) for this text: '{text}'"}
            ]
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return "Error"

# Function for Together sentiment analysis
def together_sentiment(text):
    try:
        response = together_client.chat.completions.create(
            model="meta-llama/Meta-Llama-3.1-8B-Instruct-Turbo",
            messages=[{"role": "user", "content": f"Provide the sentiment (positive, negative, neutral) for this text: '{text}'"}]
        )
        return response.choices[0].message["content"].strip()
    except Exception as e:
        return "Error"

# Predict sentiments using both APIs
data["openai_sentiment"] = data["conversation"].apply(openai_sentiment)
data["together_sentiment"] = data["conversation"].apply(together_sentiment)

# Classification reports
print("\nClassification Report for OpenAI:")
print(classification_report(data["customer_sentiment"], data["openai_sentiment"], zero_division=0))

print("\nClassification Report for Together:")
print(classification_report(data["customer_sentiment"], data["together_sentiment"], zero_division=0))

# Display data with predictions
print("\nPredictions:")
print(data)

```
