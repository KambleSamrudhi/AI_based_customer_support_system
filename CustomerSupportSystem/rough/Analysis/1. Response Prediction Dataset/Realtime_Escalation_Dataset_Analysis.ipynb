```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/drive/MyDrive/helpdesk_customer_multi_lang_tickets.csv")
```

```python
data.head()
```

```python
# 1. Message Types
message_types = data["type"].value_counts()
print("\nMessage Types:\n",message_types)

# 2. Queues
queues_distribution = data["queue"].value_counts()
print("\nQueues Dist:\n",queues_distribution)

# 3. Priority Levels
priority_distribution = data["priority"].value_counts()
print("\nPriority Dist:\n",priority_distribution)

# 4. Languages
language_distribution = data["language"].value_counts()
print("\nLanguage Dist:\n",language_distribution)

# 5. Business Types
business_distribution = data["business_type"].value_counts()
print("\nBusiness Dist:\n",business_distribution)

# Visualization

# Plot Message Types
message_types.plot(kind="bar", title="Message Type Distribution", color="skyblue")
plt.xlabel("Message Type")
plt.ylabel("Frequency")
plt.show()

# Plot Queues Distribution
queues_distribution.plot(kind="pie", autopct="%1.1f%%", title="Queue Distribution", colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"])
plt.ylabel("")  # Hide y-label
plt.show()

# Plot Priority Distribution
priority_distribution.plot(kind="bar", title="Priority Distribution", color="orange")
plt.xlabel("Priority")
plt.ylabel("Frequency")
plt.show()

# Plot Languages
language_distribution.plot(kind="bar", title="Language Distribution", color="green")
plt.xlabel("Language")
plt.ylabel("Frequency")
plt.show()

# Plot Business Types
business_distribution.plot(kind="pie", autopct="%1.1f%%", title="Business Type Distribution", colors=["#c44e52", "#4c72b0"])
plt.ylabel("")  # Hide y-label
plt.show()
```
